### Lesson 17: 'For' Loops

```r
# recall in the Lesson 10 we want to replace 0 values with NA
pima$glucose[pima$glucose == 0] <- NA
sum(is.na(pima$glucose))
```

```r
# but there are several columns, we do not want to replace them line by line
# a 'for' loop can solve our problem
# !Notice: if there are NA values, most of the bool calculation would also be NA
for (i in 1:length(pima)) {
    print(sum(pima[, i] == 0))
}
```

```r
names(pima)
```

```r
# not all columns need to be replaced
# !Notice: we are interrupted when i = 2, this is because the second column contains NA values (recall that most of bool calculations of NA values are also NA)
for (i in 2:6) {
    print(i)
    pima[pima[, i] == 0, i] <- NA
}
```

```r
# however, seperate operations work
# !Notice: this means `which` operation would ignore NA values, which is different from directly writing the filter as indices
for (i in 2:6) {
    print(i)
    idx = which(pima[, i] == 0)
    pima[idx, i] <- NA
}
```

```r
# we can also do this to operate except NA values, but this is not that clean
for (i in 2:6) {
    print(i)
    pima[pima[, i] == 0 & !is.na(pima[, i]), i] <- NA
}
```

```r
# use `break` to leave the loop early
# example: adding cubes of natural numbers, and for some reason we want to end up the process after exceeding some number `s`
f <- function(n, s) {
    tot <- 0
    for (i in 1:n) {
        tot <- tot + i^3
        if (tot > s) {
            print(i)
            break
        }
        if (i == n) print('failed')
    }
}
f(100, 345)
```

```r
f(5, 345)
```

### Lesson 18: Functions with Blocks

```r
# recall the data cleaning process we did in Lesson 17. We can wrap it to be a function for convenient reuse.
# `return` can be ignored by simiply typing the variable we want to return
# use `colSums` to get the column sums
zerosToNAs <- function(d,cols) {
   for (j in cols) {
      NArows <- which(d[, j] == 0)
      d[NArows, j] <- NA
   }
   d
}
colSums(is.na(zerosToNAs(pima, 2:6)))
```

### Lesson 20: If, Else, Ifelse

```r
# author's dataset: prgeng
# show first 15 lines of `prgeng$educ`
head(prgeng$educ, 15)
```

```r
# we want the `educ` code to be at least 12
# !Notice: we can see from the cell above that `educ` is a factor instead of numeric, and comparing a factor with numeric is meaningless, so we need to transform it to numeric first
prgeng$educ <- as.numeric(prgeng$educ)
for (i in 1:nrow(prgeng)) {
    if (prgeng$educ[i] < 13) prgeng$educ[i] <- 12
}
head(prgeng$educ, 15)
```

```r
# there is a function `ifelse` can do the same thing simpler
# it works on vectors
educ <- as.numeric(prgeng$educ)
prgeng$educ <- ifelse(educ < 13, 12, educ)  # (test, if TRUE, else)
head(prgeng$educ, 15)
```

```r
# now we can plot w.r.t `educ`
plot(prgeng$age, prgeng$wageinc, col=educ)
```

```r
# try again: Nile
nile <- ifelse(Nile > 1150, 3, 2)
nile <- ifelse(Nile < 800, 1, nile)
table(nile)
```

### Lesson 21: Do Professional Athletes Keep Fit?

```r
# author's dataset: mlb
# Many people gain weight as they age. But what about professional athletes?
options(timeout=600)
load(url('https://github.com/matloff/fasteR/blob/master/data/mlb.RData?raw=true'))
head(mlb)
```

```r
# we can observe that the `Age` are float and we want to make them integer for convenience to group
# using `round` function
age <- round(mlb$Age)
table(age)
```

```r
# now we want to get the group means of `Weight` to see whether athletes gain weight as they age
taout <- tapply(mlb$Weight, age, mean)
taout
```

```r
# note that the above cell output do not give the usual index instruction like [1], [16], etc.
# we can get that instructions by calling `names` function
names(taout)
```

```r
# now we can see the relationship between age and weight
# result shows that professional athletes also gain weight as they age
plot(23:35, taout[3:15])
```

### Lesson 22: Linear Regression Analysis, I

```r
# in R, linear model can be created by `lm` function
lm(Weight ~ Age, data=mlb)  # just write expressions using column names without data frame name
```

```r
# we can add a fitted line on our scatter plot now
plot(23:35, taout[3:15])
abline(181.4366, 0.6936)  # (intercept, slope)
```

### Lesson 23: S3 Classes

```r
# we don't want to type the intercept and slope by hand like what we did in Lesson 22
# here is a more propriate way to do that
# we can see the class of the output of `lm` first -- just an `lm` class
lmout <- lm(Weight ~ Age, data=mlb)
class(lmout)
```

```r
# a handy way to take a quick glance at the contents of an object is `str`
str(lmout)
```

```r
# from the output above we know that `lm` is also a list, so we can get `coefficients` directly using `$`
# `coefficients` is a numeric vector
cfs <- lmout$coefficients
class(cfs)
```

```r
# now we can add a fitted line without handy inputs
plot(23:35, taout[3:15])
abline(cfs[1], cfs[2])  # (intercept, slope)
```

```r
# we can also do a multiple regression
lmo <- lm(Weight ~ Height + Age, data=mlb)
lmo
```

### Lesson 24: Baseball Player Analysis (cont'd.)

```r
# we want to know whether the regression model is different among player positions
table(mlb$PosCategory)
```

```r
# so we want to split the data frame according to `PosCategory`
# however, `split` function works only for vectors (not for data frame)
# we use an indirect way: split the row indices
rownums <- split(1:nrow(mlb), mlb$PosCategory)
rownums
```

```r
class(names(rownums))
```

```r
# now we create an empty data frame and put the regression coefficients of each positon in it
# `data.frame()` function can create a data frame
# `rbind` (row bind) function is like `append` in python
posNames <- names(rownums)
m <- data.frame()
for (pos in posNames) {
   posRows <- rownums[[pos]]
   lmo <- lm(Weight ~ Age, data=mlb[posRows, ])
   newrow <- lmo$coefficients
   m <- rbind(m, newrow)
}
m
```

```r
# the data frame we form is ugly, we can replace the row index with `posNames` and set the column names
row.names(m) <- posNames
names(m) <- c('intercept', 'slope')
m
```
